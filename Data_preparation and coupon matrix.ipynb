```python
# Allow FULL_DATE to be before Start Date (negative days are valid)
merged_df['days_since_start'] = (merged_df['FULL_DATE'] - merged_df['Start Date']).dt.days

# Bin into 2-week periods (each bin = 14 days), allowing negative bins
merged_df['week_bin'] = (merged_df['days_since_start'] // 14) * 2

# Keep only the 36 bins: from -26 to +50 (13 pre-trial, 13 in-trial, 13 post-trial)
data = merged_df[merged_df['week_bin'].between(-26, 50)]

# Label phase based on week_bin
def label_phase(x):
    if x <= -2:
        return 'Pre-Trial'
    elif 0 <= x <= 24:
        return 'In-Trial'
    elif x >= 26:
        return 'Post-Trial'
    else:
        return None  # week_bin == -0 or 23 are edge cases and excluded by design

merged_df['phase'] = merged_df['week_bin'].apply(label_phase)
data = merged_df[merged_df['week_bin'].between(-26, 50)]
merged_df
```

```python
# Find the dropped rows by comparing indexes
dropped_rows = merged_df[~merged_df.index.isin(data.index)]
dropped_rows
# Summarize unique PIDs and their associated Intervention and phase
drop_summary = dropped_rows.groupby(['PID','COHORT #', 'Intervention',])['phase'].first().reset_index()

drop_summary
```
